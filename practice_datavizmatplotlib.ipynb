```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['Species'] = iris.target
df
```

```python
print(df.info())
print(df.describe())
```

```python
features = iris.data.T
Labels = ['Setosa','Versicolour','Virginica']
colors = ['r','g','b']
classlbl = iris.target

fig = plt.figure(figsize = (12, 10))
ax = fig.add_subplot(111)

colors = np.asarray(colors)
colorslist = colors[classlbl]
Labels = np.asarray(Labels)
labellist = Labels[classlbl]

for x,y,s,c,l in zip(features[0], features[1], features[3], colorslist, labellist):
    ax.scatter(x, y, s=100*s, color = c, label = 1, cmap = 'viridis', alpha = 0.2)
    
handles, labels = ax.get_legend_handles_labels()

newLeg = dict()

for h, l in zip(handles, labels):
    if 1 not in newLeg.keys():
        newLeg[1] = h

# new handles + labels
handles = []
labels = []
for l in newLeg.keys():
    handles.append(newLeg[1])
    labels.append(1)

# new legend
ax.legend(handles, labels, loc='lower right', frameon = True, prop = {'size': 20})
ax.set_xlabel(iris.feature_names[0], fontsize = '22')
ax.set_ylabel(iris.feature_names[1], fontsize = '22')
ax.set_title('Demonstration of Data Viz', fontsize = '22')
plt.show()
```
